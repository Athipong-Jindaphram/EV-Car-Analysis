```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv("/Users/athipongjindaphram/Documents/EV Car Analysis/ev_market_2026.csv")
df.sample()
```

```python
df_drive_type = df.copy()
df_drive_type['efficiency'] = df_drive_type['range_miles'] / df_drive_type['battery_capacity_kwh']

plt.figure(figsize=(7, 6))
sns.violinplot(x='drive_type', y='efficiency', data=df_drive_type)

plt.title('EV Efficiency (Miles per kWh) by Drive Type')
plt.ylabel('Efficiency (Miles / kWh)')
plt.xlabel('Drive Type')
sns.despine()
plt.grid(alpha=0.3)
plt.show()
```

```python
g = sns.FacetGrid(df, col="market_segment", height=4)
g.map(sns.histplot, "safety_rating", bins=[3, 4, 5, 6], discrete=True)

plt.show()
```

```python
autopilot_pct = pd.crosstab(df['market_segment'], df['autopilot_level'], normalize='index') * 100

autopilot_pct.plot(kind='barh', stacked=True, figsize=(12, 6), colormap='coolwarm')
plt.title('Autopilot Technology Adoption (Percentage by Segment)', fontsize=14)
plt.xlabel('Percentage (%)')
plt.legend(title='Autopilot Level', bbox_to_anchor=(1, 1))
plt.grid(alpha=0.5)
plt.show()
```
