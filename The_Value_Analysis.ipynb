```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv("/Users/athipongjindaphram/Documents/EV Car Analysis/ev_market_2026.csv")
df.sample()
```

```python
plt.scatter(df["price_usd"], df["range_miles"], alpha=0.2, color="red")

plt.title("Price for Range")
plt.ylabel("Range (Miles)")
plt.xlabel("Price ($USD)")
plt.tight_layout()
plt.show()
```

```python
df_body_type = df.copy()

df_body_type = df_body_type.groupby("body_type")["price_usd"].mean().sort_values(ascending=False)

df_body_type.plot(kind="bar")

sns.despine()
sns.set_theme(style="ticks")

plt.grid(axis="y", alpha=0.7, linestyle="--")
plt.title("Average Price by Body Type")
plt.xlabel("Body Type")
plt.ylabel("Average Cost ($USD)")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
plt.scatter(df['price_usd'], df['customer_rating'], alpha=0.3, color="green")

sns.set_theme(style="ticks")

plt.title("Customer Satisfaction vs. Pricing")
plt.xlabel("Price ($USD)")
plt.ylabel("Rating")
plt.tight_layout()
plt.show()
```
